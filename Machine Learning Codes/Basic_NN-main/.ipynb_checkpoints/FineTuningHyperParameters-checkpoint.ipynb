```python
import tensorflow as tf
from sklearn.datasets import fetch_california_housing

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Model

from tensorflow.keras.utils import plot_model
```

# 1 Simple GridSearch Method

```python
# load the toy data
housing = fetch_california_housing()

X_train_f, X_test, y_train_f, y_test = train_test_split(housing.data, housing.target)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_f, y_train_f)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_valid = sc.transform(X_valid)

r_val, features_ = X_train.shape
```

```python
# define a FUNCTION to create a model using FUNCTIONal API
def build_model(n_hidden=4, n_neurons=30, learning_rate=3e-3, input_shape=[8]):
  input_1 = Input(shape=input_shape, name='input_layer')
  if n_hidden==1:
    x = Dense(n_neurons, activation='relu')(input_1)
  else:
    for layer in range(n_hidden):
      name_ = f'dense_{layer+1}'
      if layer == 0:
        x = Dense(n_neurons, activation='relu', name=name_)(input_1)
      else:
        x = Dense(n_neurons, activation='relu', name=name_)(x)
  output_ = Dense(1, name='output_layer')(x)

  optimizer = Adam(learning_rate=learning_rate)

  model = Model(inputs=[input_1], outputs=[output_])

  model.compile(loss='mse', optimizer=Adam(learning_rate=learning_rate))
  return model

model = build_model(n_hidden=3)

# check if the model is as expected.
plot_model(model, show_shapes=True, show_layer_activations=True)

# or
# model.summary()
```

```python
# compile and fit the model
history = model.fit(X_train, y_train, epochs=30, validation_data=(X_valid, y_valid), verbose=4)
```

```python
# create a KerasRegressor
keras_reg = tf.keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

```python
keras_reg.fit(X_train, y_train, epochs=100,
              validation_data=(X_valid, y_valid),
              callbacks=[tf.keras.callbacks.EarlyStopping(patience=10)])

mse_test = keras_reg.score(X_test, y_test)
```

```python
# lets use RandomizedSearchCV from sklearn
# for sake clearity all the imports are put below

from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
# lets define parameters ranges
# key values must be as defined below (build_model function)
#build_model(n_hidden=4, n_neurons=30, learning_rate=3e-3, input_shape=[8]) 
param_distribs = {
    "n_hidden": [1, 2, 3, 4, 5, 6],
    "n_neurons": np.arange(1, 50),
    "learning_rate": reciprocal(3e-4, 3e-2)
}

rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=10, cv=3)

rnd_search_cv.fit(X_train, y_train, epochs=100, validation_data=(X_valid, y_valid),
                  callbacks=[tf.keras.callbacks.EarlyStopping(patience=10)])
```

```python
rnd_search_cv.best_params_
```

```python
rnd_search_cv.best_score_
```

```python
model = rnd_search_cv.best_estimator_.model
```

```python
model.save('Fine_tune')
```

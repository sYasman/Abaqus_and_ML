```python
import tensorflow as tf
import tensorflow_datasets as tfds

from tensorflow.keras.utils import plot_model
import matplotlib.pyplot as plt
```

```python
(train_data, valid, test), info = tfds.load('cifar10', 
                                split=['train[:75%]', 'train[75%:]', 'test'],
                                with_info=True, as_supervised=False)
```

```python
fig = tfds.show_examples(train_data, info)
```

```python
print(info)
```

```python
def normalize_image(data):
  image = data["image"]
  label = data["label"]

  image = tf.cast(image, 'float32') / 255.
  return image, label
```

```python
BUFFER_SIZE = 1024
BATCH_SIZE = 32

train_norm = train_data.map(normalize_image).shuffle(BUFFER_SIZE).batch(BATCH_SIZE).prefetch(1)
valid_norm = valid.map(normalize_image).shuffle(BUFFER_SIZE).batch(BATCH_SIZE).prefetch(1)
test_norm = test.map(normalize_image).batch(BATCH_SIZE).prefetch(1)
```

```python
# for e in train_norm.take(1):
#   print(e)
```

```python
#using Functional API create a simple model.
input_ = tf.keras.Input(shape=(32, 32, 3), name='input_')
x = tf.keras.layers.Conv2D(64, 3, activation="relu", padding="same")(input_)
x = tf.keras.layers.MaxPooling2D(2)(x)
x = tf.keras.layers.Conv2D(64, 3, activation="relu", padding="same")(x)
x = tf.keras.layers.Conv2D(64, 3, activation="relu", padding="same")(x)
x = tf.keras.layers.MaxPooling2D(2)(x)
x = tf.keras.layers.Conv2D(64, 3, activation="relu", padding="same")(x)
x = tf.keras.layers.Conv2D(64, 3, activation="relu", padding="same")(x)
x = tf.keras.layers.MaxPooling2D(2)(x)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)
x = tf.keras.layers.Dropout(0.50)(x)
output_ = tf.keras.layers.Dense(10, activation='softmax')(x)

model = tf.keras.Model(inputs=input_, outputs=output_)
```

```python
plot_model(model, show_shapes=True, show_layer_names=True, show_layer_activations=True)
```

```python
optimizer_ = tf.keras.optimizers.Adam()
loss_ = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)
# Since activation is softmax. There is no need to from_logits true.
# (https://stackabuse.com/what-is-fromlogitstrue-in-keras-tensorflow-loss-functions/)

tr_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy()
cb_early_stop = tf.keras.callbacks.EarlyStopping(patience=5)
```

```python
# Results for NORMALIZED DATA
# model w/o normlization
model.compile(optimizer=optimizer_, loss=loss_, metrics=tr_acc_metric)
h = model.fit(train_norm, epochs=50, validation_data=valid_norm, callbacks=[cb_early_stop])
```

```python
plt.figure(figsize=(10, 7))
for k in h.history.keys():
  plt.plot(h.history[str(k)], label=str(k))
plt.legend()
```

```python
model.save('cifar-10.h5')
```

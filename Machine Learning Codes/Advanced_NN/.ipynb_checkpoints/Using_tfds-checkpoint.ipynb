# TensorFlow Datasets

```python
from tensorflow._api.v2.data import AUTOTUNE
import matplotlib.pyplot

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow_datasets as tfds

BATCH_SIZE = 64
```

## Introductroy Example (cifar10)

```python
(ds_train, ds_test), ds_info = tfds.load(
    "cifar10",
    split=["train", "test"],
    as_supervised=True,
    shuffle_files=True,
    with_info=True
)
```

```python
# lets see the details of dataset
print(ds_info)
```

```python
# tfds.show_examples(ds_train, ds_info)
```

```python
# function to normalize images
def normalize_img(img, label):
  return tf.cast(img, tf.float32)/255.0, label
```

```python
ds_train = ds_train.map(normalize_img, num_parallel_calls=AUTOTUNE) # unless assign it to new variable it is not written to the old on
```

```python
ds_train = ds_train.cache()
ds_train = ds_train.shuffle(ds_info.splits["train"].num_examples)
ds_train = ds_train.batch(BATCH_SIZE)
ds_train = ds_train.prefetch(AUTOTUNE)
```

```python
ds_test = ds_test.map(normalize_img, num_parallel_calls=AUTOTUNE)
ds_test = ds_test.cache()
# ds_test = ds_test.shuffle(ds_info.splits["train"].num_examples) #no shuffle in test set
ds_test = ds_test.batch(BATCH_SIZE*2)
ds_test = ds_test.prefetch(AUTOTUNE)
```

```python
n_filters=32
kernel_size=3
input_ = layers.Input(shape=(32, 32, 3))
x = layers.Conv2D(filters=n_filters*2, kernel_size=3, padding='same', activation='relu')(input_)
x = layers.MaxPool2D()(x)
x = layers.Conv2D(filters=n_filters, kernel_size=3, padding='same', activation='relu')(x)
x = layers.Conv2D(filters=n_filters, kernel_size=3, padding='same', activation='relu')(x)
x = layers.Conv2D(filters=n_filters, kernel_size=3, padding='same', activation='relu')(x)
x = layers.Flatten()(x)
outputs_ = layers.Dense(10, activation='softmax')(x)

model = keras.Model(inputs=input_, outputs=outputs_)
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss=keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"]
)

hist = model.fit(ds_train, epochs=50, batch_size=64)
eval = model.evaluate(ds_test)
```

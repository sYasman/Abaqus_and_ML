```python
import tensorflow as tf
import tensorflow_datasets as tfds

from tensorflow.keras.utils import plot_model
```

```python
# load and preprocess data
X_train, info = tfds.load("cifar10",
                          split='train',
                          with_info=True)

X_test = tfds.load("cifar10",
                          split='test',
                          with_info=False)
```

```python
info
```

```python
fig = tfds.show_examples(X_train, info)
```

```python
X_test.take(1)
```

```python
# format data
# you may add augmentation as well here
def format_image(data):
  image = data['image']
  label = data['label']

  image = tf.cast(image, 'float32')
  image = image / 255.

  return image, label
```

```python
BUFFER_SIZE = 1024
BATCH_SIZE = 64

X_tr_norm = X_train.map(
    format_image, num_parallel_calls=tf.data.AUTOTUNE)
X_tr_norm = X_tr_norm.shuffle(BUFFER_SIZE).batch(BATCH_SIZE)
X_tr_norm = X_tr_norm.cache()
X_tr_norm = X_tr_norm.prefetch(tf.data.AUTOTUNE)

X_test_norm = X_test.map(
    format_image, num_parallel_calls=tf.data.AUTOTUNE)
X_test_norm = X_test_norm.batch(BATCH_SIZE)
X_test_norm = X_test_norm.cache()
X_test_norm = X_test_norm.prefetch(tf.data.AUTOTUNE)
```

```python
# define model
def base_model():
  input_ = tf.keras.Input(shape=(32, 32, 3), name='input')
  x = tf.keras.layers.Conv2D(32, (3, 3), activation='relu',\
                              kernel_initializer='he_uniform', padding='same')(input_)
  x = tf.keras.layers.MaxPooling2D()(x)
  x = tf.keras.layers.Conv2D(32, (3, 3), activation='relu',\
                              kernel_initializer='he_uniform', padding='same')(x)
  x = tf.keras.layers.MaxPooling2D()(x)

  x = tf.keras.layers.Flatten()(x)
  x = tf.keras.layers.Dense(64, activation='relu')(x)
  
  output_ = tf.keras.layers.Dense(10, activation='softmax', name='pred')(x)

  model = tf.keras.Model(inputs=input_, outputs=output_)
  return model
```

```python
model = base_model()

plot_model(model, show_shapes=True, show_layer_names=True, show_layer_activations=True)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-4)
loss_object = tf.keras.losses.SparseCategoricalCrossentropy()

train_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy()
val_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy()
```

```python
model
```

```python
model.compile(optimizer=optimizer, loss=loss_object, metrics=train_acc_metric)
hist = model.fit(X_tr_norm, epochs=50, validation_data=X_test_norm)
```

# Create a Custom Training Loop

 Our objective is minimizim loss function.
 Lets assume we are using standard Newton Raphson Algorithm for one variable. If this is the case, we must compute the derivative of loss function, then using this derivative we update all the other variables. Basically,

 $$x_{n+1} = x_{n} - \dfrac{f(x)}{f'(x)}$$

 Thus, with the following function, we simply compute the gradients and loss function and appy them to the solution algorithm. Since we are updating weights regarding the loss function, the update algorithm takes the following form,

 $$\boldsymbol{W_{n+1}} = \boldsymbol{W_{n+1}} - 𝕃 \left( ∇_{\boldsymbol{W}}𝕃 \right)^{-1}$$

```python
def apply_gradient(optimizer, x, y):
  '''
  Funtion to provide the gradients of loss function to the optimizer.
  optimizer: tf.keras.optmizers.__
  x = Training features (batched)
  y = Trainin labels (batched)
  '''
  with tf.GradientTape() as tape:
    logits = model(x)
    loss_value = loss_object(y_true=y, y_pred=logits)
  gradients = tape.gradient(loss_value, model.trainable_weights)
  optimizer.apply_gradients(zip(gradients, model.trainable_weights))

  return logits, loss_value
```

Now for each bath, the new weights must be computed. Furthermore, loss and accuracy must be updated. For this purpose the function below will be utilized.

```python
def train_data_for_one_epoch(optimizer, X_tr_norm):
  losses = []

  for step, (x_batch_tr, y_batch_tr) in enumerate(X_tr_norm):
    # update gradients, loss, logits
    logits, loss_value = apply_gradient(optimizer, x_batch_tr, y_batch_tr)
    losses.append(loss_value)
    # update accuracy
    train_acc_metric(y_batch_tr, logits)
  return losses
```

At the end of each epoch, one has to validate the model. For this purpose the test set will be used with the following function.

```python
def perform_validation(X_test_norm):
  losses = []
  for x_val, y_val in X_test_norm:
    val_logits = model(x_val)
    val_loss = loss_object(y_val, val_logits)

    losses.append(val_loss)

    val_acc_metric(y_val, val_logits)
  return losses
```

Now lets define the training loop. In the training loop, for a fixed number of epochs, loss is minimized. The training loop must consist of,


1.   Perform training over the batches of training data
2.   Get metrics
3.   Perform validation to calculate loss and update validation metrics on data
4.   Reset metrics at the end of the epoch
5.   Display statistics at the end of each epoch


```python
import numpy as np

model = base_model()

# Iterate over epochs.
epochs = 20
epochs_val_losses, epochs_train_losses = [], []
epochs_val_acc, epoch_train_acc = [], []
for epoch in range(1, epochs+1):
  print('\nStart of epoch %d' % (epoch,))
  
  losses_train = train_data_for_one_epoch(optimizer, X_tr_norm)
  train_acc = train_acc_metric.result()

  losses_val = perform_validation(X_test_norm)
  val_acc = val_acc_metric.result()

  losses_train_mean = np.mean(losses_train)
  losses_val_mean = np.mean(losses_val)
  epochs_val_losses.append(losses_val_mean)
  epochs_train_losses.append(losses_train_mean)

  epochs_val_acc.append(val_acc)
  epoch_train_acc.append(train_acc)  

  print('\n Epoch %s: Train loss: %.4f  Validation Loss: %.4f, Train Accuracy: \
  %.4f, Validation Accuracy %.4f' % (epoch, float(losses_train_mean), \
                                     float(losses_val_mean), \
                                     float(train_acc), \
                                     float(val_acc)))
  
  train_acc_metric.reset_states()
  val_acc_metric.reset_states()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

def plot_metrics(train_metric, val_metric, metric_name, title, ylim=(0, 1)):
  plt.figure(figsize=(10, 6))
  plt.title(title)
  plt.ylim((ylim))
  plt.gca().xaxis.set_major_locator(mticker.MultipleLocator(1))
  plt.plot(train_metric,color='blue',label=metric_name)
  plt.plot(val_metric,color='green',label='val_' + metric_name)
  plt.legend()
  plt.grid()

plot_metrics(epochs_train_losses, epochs_val_losses, "Loss", "Loss", ylim=(0, 2))
plot_metrics(epoch_train_acc, epochs_val_acc, "Acc", "Acc", ylim=(0.40,1))
```
